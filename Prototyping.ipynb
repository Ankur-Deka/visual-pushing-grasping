Make trained model compatible with newer versions of torch

```python
import torch
import collections
# print(model.keys())a
# print(model['push_color_trunk.features.denseblock1.denselayer1.norm.1.weight'])
model = torch.load('downloads/vpg-original-sim-pretrained-10-obj-original.pth')
new_model = collections.OrderedDict()

for key in model.keys():
    new_key = key.replace('norm.1', 'norm1')
    new_key = new_key.replace('norm.2', 'norm2')
    new_key = new_key.replace('norm.3', 'norm3')
    new_key = new_key.replace('norm.4', 'norm4')
    new_key = new_key.replace('conv.1', 'conv1')
    new_key = new_key.replace('conv.2', 'conv2')
    new_key = new_key.replace('conv.3', 'conv3')
    new_key = new_key.replace('conv.4', 'conv4')
    print('key', key)
    print('new key', new_key)
    new_model[new_key] = model[key]

torch.save(new_model, 'downloads/vpg-original-sim-pretrained-10-obj.pth')
print(new_model['push_color_trunk.features.transition1.norm.running_mean'])
    
```
